```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
X = iris.data[:, [2,3]] #All rows, 3rd and 4th columns.
y = iris.target
```

To get started with the scikit-learn Library, we will train a perception model similar to the one implemented in Chapter 2 using the Iris Dataset.


```python
import pandas as pd
iris2 = datasets.load_iris(as_frame=True)
iris = iris2.frame
print(iris.head())
```

```python
print('Class Labels:', np.unique(y))
```

The np.unique() function returns the three unique class labels that are stored in iris.target. The Iris flower class names are stored as integers here. Although many sci-kit learn functions work with class labels in string format, integer labels are preferred to avoid issues.

How to evaluate how well the trained model performs on unseen data? Train-test split.


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

The train_test_split funciton already shuffles the training datasets internally before splitting. random_state is a random seed for the internal pseudo-random number generator. Using a fixed number means our results are reproducible.

Min-Max Normalization rescales the data while retaining the relative positions of the data.

Z-score normalization takes the square of the difference between each value and the mean of all of the values, divides by the number of values then square roots the entire calculation. This calculates the standard deviation. Then x-mu/sigma to normalize the values.

The transformed units are dimensionless




```python
#Let's standardize the features by scaling them for optimal performance.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

#the fit method of the StandardScaler object estimates the parameters mu and sigma. We then call the transform method to standardize the data using those values.
#Having standardized the training data, we can now train a perception model. We can feed all three flower classes to the perception all at once.

from sklearn.linear_model import Perceptron
ppn = Perceptron(eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)

#y_train contains the labels. The fit method trains the model. eta is the learning rate.

y_pred = ppn.predict(X_test_std)
print('Misclassified examples: %d' % (y_test != y_pred).sum())
```

```python
#The perceptron misclassifies 1 out of the 45 flower examples, so the misclassification error is about 2.2%
from sklearn.metrics import accuracy_score
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))

#Let's use plot_decision_regions from ch 2 to plot the decision regions of the newly trained perceptron.
#Visualizes the separations.

from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
  #"setup marker generator and color map" (?)
  markers = ('o','s', '^', 'v', '<')
  colors = ('red', 'blue' ,'lightgreen', 'gray', 'cyan')
  cmap = ListedColormap(colors[:len(np.unique(y))])
  #plotting the decision surface
  x1_min, x1_max = X[:,0].min()-1, X[:,0].max() + 1 #Calculate minimum and maximum values of the first feature
  x2_min, x2_max = X[:,1].min()-1, X[:,1].max()+1   #Calculate minimum and maximum values of the second feature
  xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),  #Creates a 2d mesh grid with step size 'resolution', in this case along both dimensions.
                         np.arange(x2_min, x2_max, resolution))  #xx1 and xx2 are the x coordinates and the y coordinates respectively for each point in the mesh.
  lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T) #ravel converts the 2d arrays into 1d arrays. then we create an array consisting of coordinate pairs. We put the coordinate pairs into the
  #classifier, making a prediction for every pair of coordinates in the mesh.
  lab = lab.reshape(xx1.shape) #We reshape lab back into the same shape as xx1.
  plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)
  plt.xlim(xx1.min(), xx1.max())
  plt.ylim(xx2.min(), xx2.max())

  for index, cl in enumerate(np.unique(y)):
    plt.scatter(x=X[y== cl, 0],
                y=X[y == cl, 1],
                alpha=0.8,
                c=colors[index],
                marker=markers[index],
                label=f'class {cl}',
                edgecolor = 'black')
  if test_idx:
    #plot all examples
    X_test, y_test = X[test_idx, :], y[test_idx]

    plt.scatter(X_test[:, 0], X_test[:,1],
                c='None', edgecolor='black', alpha=1.0,
                linewidth=1, marker='o',
                s=100, label='Test Set')



```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std,
                      y=y_combined,
                      classifier=ppn,
                      test_idx=range(105,150))
plt.xlabel('Standardized Petal length')
plt.ylabel('Standardized Petal width')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

The three flower classes cannot be perfectly separated by a linear decision boundary.....The perceptron algorithm will not converge on datasets that are not perfectly linearly separable.
There are more powerful linear classifiers that will converge to a loss minimum even if the classes are not perfectly linearly separable.
